## Seaborn Chart

```python
import numpy as np
import seaborn as sns
sns.set(style="white")

rs = np.random.RandomState(10)
d = rs.normal(size=100)

sns.histplot(d, kde=True, color="b", bins=20, stat="density")
```

```python
import seaborn as sns

sns.set(style="ticks")

df = sns.load_dataset("anscombe")

sns.lmplot(x="x",y="y",data=df)
```
